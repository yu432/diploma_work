```python
def split_into_chapters(filename):
    separators = ["I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX", "X",
                  "XI", "XII", "XIII", "XIV", "XV", "XVI", "XVII", "XVIII", "XIX", "XX",
                  "XXI", "XXII", "XXIII", "XXIV", "XXV", "XXVI", "XXVII", "XXVIII"]
    
    chapters = []  
    chapter_text = []  

    with open(filename, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    lines.append("XXVIII конец")

    for line in lines:
        line_stripped = line.strip()
        if any(line_stripped.startswith(sep) for sep in separators):
            if chapter_text:  
                chapter_content = ''.join(chapter_text).strip()
                words = chapter_content.split()
                word_count = len(words)
                
                if word_count > 2500:
                
                    mid_point = word_count // 2
                    part1 = ' '.join(words[:mid_point])
                    part2 = ' '.join(words[mid_point:])
                    chapters.append((part1, len(part1.split())))
                    chapters.append((part2, len(part2.split())))
                else:
                    chapters.append((chapter_content, word_count))
                
                chapter_text = [] 
        chapter_text.append(line)  

    return chapters

chapters_list = split_into_chapters('diploma_work/book_level_long/turgenev_ottsy_i_deti.txt')
for index, (chapter, word_count) in enumerate(chapters_list):
    print(f"Глава {index+1}: количество слов - {word_count}")
```

```python
import gc
import torch

def summarize_chapters(chapters):
    summaries = []  

    for index, (chapter_content, _) in enumerate(chapters):
        print("идет суммаризация главы:")
        print(index)
        torch.cuda.empty_cache()
        gc.collect()

        messages = [
            {"role": "system", "content": "Ты помощник, отвечай на следующий текст по-русски"},
            {"role": "user", "content": f'Напиши суммаризацию этого текста, пиши по-русски, используй не больше 300 слов:\n{chapter_content}'}
        ]

        prompt = pipeline.tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True
        )

        print(len(pipeline.tokenizer(messages[1]['content'])['input_ids']))

        
        terminators = [
          pipeline.tokenizer.eos_token_id,
          pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
        ]

        
        outputs = pipeline(
            prompt,
            max_new_tokens=768,
            eos_token_id=terminators,
            do_sample=True,
            temperature=0.25,
            top_k=100,
            top_p=0.98,
        )

        summary_text = outputs[0]["generated_text"][len(prompt):]
        summaries.append(summary_text)

    return summaries

summaries_list = summarize_chapters(chapters_list)
```
